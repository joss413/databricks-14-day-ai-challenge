```python
# Load data
events = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv", header=True, inferSchema=True)
display(events.limit(10))
```

```python
events.columns
```

```python
# select operation
events.select("event_time", "category_code", "price").show(10)
```

```python
# filter operation
from pyspark.sql.functions import col
filtered_events = events.filter(col("price") > 100)
filtered_events.select("brand", "price").show()
```

```python
# group by operation
events.groupBy("brand").count().show()
top_brands = events.groupBy("brand").count().orderBy("count", ascending=False).limit(5)
top_brands.show()
```

```python
display(spark.sql("SHOW VOLUMES"))
```

```python
# Export as a table
top_brands.write.mode("overwrite").saveAsTable("top_brands_results")
print("✅ Saved as table: top_brands_results")

# Display it
display(spark.sql("SELECT * FROM top_brands_results"))
```
