```python

# 1. Add widgets for parameters
dbutils.widgets.text("source_path", "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv")
dbutils.widgets.dropdown("layer", "bronze", ["bronze", "silver", "gold"])

# 2. Use parameters
source = dbutils.widgets.get("source_path")
layer = dbutils.widgets.get("layer")

print(f"Running {layer} layer")
print(f"Source: {source}")

# 3. Define run_layer function
def run_layer(layer_name):
    base_path = "/Volumes/workspace/ecommerce/ecommerce_data/workflow"
    
    if layer_name == "bronze":
        # Bronze logic
        print("Running Bronze logic...")
        df = spark.read.csv(source, header=True, inferSchema=True)
        df.write.format("delta").mode("overwrite").save(f"{base_path}/bronze")
        print("✓ Bronze completed")
        
    elif layer_name == "silver":
        # Silver logic
        print("Running Silver logic...")
        df = spark.read.format("delta").load(f"{base_path}/bronze")
        df.write.format("delta").mode("overwrite").save(f"{base_path}/silver")
        print("✓ Silver completed")
        
    elif layer_name == "gold":
        # Gold logic
        print("Running Gold logic...")
        df = spark.read.format("delta").load(f"{base_path}/silver")
        df.write.format("delta").mode("overwrite").save(f"{base_path}/gold")
        print("✓ Gold completed")

# 4. Run the layer
run_layer(layer)

print("\n✅ Task completed!")
```
